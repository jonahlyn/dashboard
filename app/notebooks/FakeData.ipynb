```python
data = pd.DataFrame({
    'Date': date_list,
    'Interval': pd.Series(date_list).dt.strftime('%H%M').astype('int64'),
    'Cars': [(x.dayofweek + np.random.randint(-1*x.dayofweek, 10)) for x in date_list],
    'Trucks': [(x.dayofweek + np.random.randint(-1*x.dayofweek, 10)) for x in date_list]
    }
)

data['Total'] = data['Cars'] + data['Trucks']
```
