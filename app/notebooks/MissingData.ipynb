```python
from sqlalchemy import create_engine
import pandas as pd
```

```python
# Create a database connection
conn = create_engine(DATABASE_URI, echo=False)
# Query the database
query = "SELECT date, vehicles FROM data where date between '2019-01-06' and '2019-01-07'"
df = pd.read_sql(query, conn)

# Create a full series of datetime values
idx = pd.date_range('2019-01-06', '2019-01-07', freq='5min')
df = df.groupby(pd.Grouper(key='date', freq='5min')).sum()
df.index = pd.DatetimeIndex(df.index)
df = df.reindex(idx, fill_value=0)
df = df.rename_axis('date').reset_index()

df
```
